```python
import pandas as pd
```

```python
# Regrouping the 3 data files
data_files = [
    "data/most-popular_country_weekly(Top 10).csv", 
    "data/most-popular_global_alltime 3(Top 10).csv",
    "data/most-popular_global_weekly(Top 10).csv"
    ]

# Concatenating and reading them for decoding Swedish letters/torubleshooted with LLM
df = pd.concat([pd.read_csv(file, sep=';', encoding='cp1252', on_bad_lines='warn') for file in data_files],
                ignore_index=True, sort=False)

# Returns a tuple with the total numbers of (rows, columns)
print(df.shape)

# Checking the columns name
print(df.columns)
```
